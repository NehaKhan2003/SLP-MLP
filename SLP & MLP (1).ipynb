```python
#Import required library
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Load the Dataset
url='https://raw.githubusercontent.com/ramar92/Deep-Learning-Datasets/main/Telco-Customer-Churn.csv'
data=pd.read_csv(url)
```

```python
data
```

```python
#Checking for missing value
print(data.isnull().sum())
```

```python
#Encode all categorical column using LabelEncoder
for column in data.select_dtypes(include=['object']).columns:
    le=LabelEncoder()
    data[column]=le.fit_transform(data[column])
```

```python
#Seprate feature and target variables
X = data.drop(columns=['Churn'])#All column except churn
y=data['Churn']
```

```python
#Standardize the feature
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
#Splitting the data
X_train, X_test, y_train, y_test= train_test_split(X_scaled,y,test_size=0.2)
```

```python
#Print shape to verify input dimension
print("Training input shape: ",X_train.shape)
```

Building Single Layer Perceptron

```python
slp=Sequential()
slp.add(Dense(1,activation='sigmoid',input_shape=(X_train.shape[1],)))
```

```python
#compile the model with loss,optimizer and evaluation metrics
slp.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Train the model for 30 epochs
slp.fit(X_train,y_train,epochs=30)
```

```python
#Predict on test data
y_pred=slp.predict(X_test)
y_pred=(y_pred > 0.5).astype("int")

#Calculating Accuracy
print("SLP Accuracy: ",accuracy_score(y_pred,y_test))
```

#Building MLP

```python
mlp=Sequential()
mlp.add(Dense(64,activation='relu', input_shape=(X_train.shape[1],)))#first hidden laye
mlp.add(Dense(32,activation='relu'))#Second hidden layer
mlp.add(Dense(16,activation='relu'))#Third hidden layer
mlp.add(Dense(1,activation='sigmoid'))#Output layer
```

```python
#Compile the model
mlp.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Train the model
mlp.fit(X_train, y_train, epochs=20)
```

```python
#Predict on test data
y_pred=mlp.predict(X_test)
y_pred=(y_pred > 0.5).astype("int")

#Calculating Accuracy
print("MLP Accuracy: ",accuracy_score(y_pred,y_test))
```
